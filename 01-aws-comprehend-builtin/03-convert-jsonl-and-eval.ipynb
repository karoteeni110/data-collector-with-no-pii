# Convert the results from 02.ipynb and evaluate with `eval-tab.py`

## Convert jsonl to json

```python
import json

def format_line_data(line_data):
    ((doc_id, entities),) = line_data.items()
    lst_tuples = [[entity["BeginOffset"],entity["EndOffset"]] for entity in entities]
    return {doc_id: lst_tuples}

input_file = 'results/02_results_with_score.jsonl'
output_file = 'results/02_results_to_eval.json'

output_data = {}
with open(input_file, 'r') as input_f:
    for line in input_f:
        line_data = json.loads(line)
        formatted_line_data = format_line_data(line_data)
        # print(formatted_line_data)
        output_data.update(formatted_line_data)

with open(output_file, 'w') as output_f:
    json.dump(output_data, output_f, indent=2)

print(f"{input_file} converted to JSON format and saved to {output_file}")
```
